```python
#import and set up pandas
import pandas as pd
pd.set_option("max_columns", None)
pd.set_option("max_rows", None)

#reads in data and prepares it for editing
data = pd.read_csv(r"C:/Users/Kierra/OneDrive/TMDB Box Office - Tech Challenge/train.csv")
data = data.copy();

# start converting genres into new columns
tempList = data["genres"].tolist()


##############################################################################################################
### This section is where the code would need to be changed if you wanted to reuse it for other columns
### Also, the code could probably be improved but it does work
### In short it takes all the names out of the column that we want to make into new columns
##############################################################################################################
empty = []
for i in range(0, len(tempList)):
    # takes out the front and end brackets to make the data consistent
    if(type(tempList[i]) == str): # checks if current genre isnt empty
        tempList[i] = tempList[i].replace("[", "")
        tempList[i] = tempList[i].replace("]", "")

        #splits on commas to create easily indexable strings
        tempList[i] = tempList[i].split(",")
    else:
        #if a row is empty, record its location so we dont look at it again
        empty.append(i)

# finds the strings that include the genre names
for i in range(0, len(tempList)):
    temp = []
    if(not(i in empty)): #if row isnt empty
        #find which strings include the name and adds them to a list
        for x in range (0,len(tempList[i])):
            if('name' in tempList[i][x]):
                temp.append(tempList[i][x])
               
        # for every string that included a name, remove every part that is not the actual name
        # this should leave you with something like 'Action' for example
        for x in range (0,len(temp)):
            temp[x] = temp[x].replace(" 'name': '","")
            temp[x] = temp[x].replace("'}","")

        ##############################################################################################################
        ### This is where the code is good again and is reusable as is.
        ##############################################################################################################

        #checks if this name is already a column
        # if it is then it makes the value in the current row 1
        # if not it adds a new column to the dataframe and makes the value in the current row one. All other values are zero
        for x in range (0,len(temp)):
            if temp[x] in data.columns:
                tempData = data[temp[x]].copy()
                tempData[i] = 1
                data[temp[x]] = tempData
            else:
                temp2 = [0] * len(tempList)
                temp2[i] = 1
                data[temp[x]] = temp2

data.head(10)
```
